### Step 0

Run the necessary imports.

```python
import cv2
import dlib
import matplotlib.pyplot as plt
import random
```

### Step 1

Load an image (any image that contains faces).

```python
img = cv2.imread('faces2.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Step 2

Load the dlib face predictor.

```python
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
faces = detector(img_rgb, 1)
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
img_result = img_rgb.copy()

for i, face in enumerate(faces):
    x1 = face.left()
    y1 = face.top()
    x2 = face.right()
    y2 = face.bottom()

    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    cv2.rectangle(img_result, (x1, y1), (x2, y2), color, 3)

    cv2.putText(img_result, f"Face {i+1}", (x1, y1-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(img_result)
plt.axis('off')
plt.title(f"Detected {len(faces)} faces")
plt.show()
```
